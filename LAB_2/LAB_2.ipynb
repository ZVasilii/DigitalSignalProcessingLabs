## Task 1 (Основные свойства ДВПФ)


```python
#Variant 5
N = 8
L = 3
nu_0 = 0.1

def dtft(x, M=2048):
    return  np.arange(M)/M-0.5, np.fft.fftshift(np.fft.fft(x, M))
```

### Task 1.1 ( Прямоугольный импульс в дискретной форме.)

```python
w=np.ones(N)
k = np.arange(N)
plt.figure(figsize=[8, 3])
plt.stem(k, w)
plt.title('Сигнал $w[k]$')
plt.xlabel('$k$')
plt.ylabel('$w[k]$')
plt.tight_layout()
plt.grid()
```

```python
def X(nu):
    return abs(np.sin(N*np.pi*nu) / np.sin(np.pi*nu))
M=2048
nu = np.linspace(-1,1,M)
X = [X(nu1) for nu1 in nu]

plt.figure(figsize=[8, 4])
plt.plot(nu, X)
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период(Теоретический)')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.grid()   
```

```python
plt.figure(figsize=[8, 4])
nu, W = dtft(w, M=2048)
plt.plot(nu, abs(W))
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')

#Parseval inequality
Energy = sum(w*w)
plt.xticks(np.linspace(-0.5, 0.5, 11))
plt.grid()
plt.tight_layout()
```

```python
plt.figure(figsize=[8, 4])
plt.plot(nu, np.angle(W))
plt.xlim([-0.5, 0.5])
plt.ylim([-np.pi, np.pi])
plt.title('ДВПФ w[k] (модуль), один период')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.xticks(np.linspace(-0.5, 0.5, 11))
plt.grid()
plt.tight_layout()
```

```python
print(Energy)
```

### Task 1.2 ( Прямоугольный импульс в дискретной форме.)

```python
w=np.zeros(N * L)
for i in range(w.size):
    if(i % L == 0):
        w[i] = 1.0

k = np.arange(N * L)
plt.figure(figsize=[8, 3])
plt.stem(k, w)
plt.title('Сигнал $w[k]$')
plt.xlabel('$k$')
plt.ylabel('$w[k]$')
plt.tight_layout()
plt.grid()

plt.figure(figsize=[8, 4]);
nu, W = dtft(w, M=2048)
plt.plot(nu, abs(W));
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.grid()
```

```python
def X(nu):
    return abs(np.sin(N*np.pi*nu) / np.sin(np.pi*nu))
M=2048
nu = np.linspace(-1,1,M)
X = [X(L *nu1) for nu1 in nu]

plt.figure(figsize=[8, 4])
plt.plot(nu, X)
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период(Теоретический)')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.grid()  
```

Сравнили Теоретический спектр с тем, что получили. Свойство масштабирования работает.

## Task 1.3 (Дифференцирование спектральной плотности.)

```python
w=np.ones(N)
for i in range(w.size):
        w[i] = 1.0 * i
k = np.arange(N)
plt.figure(figsize=[8, 3])
plt.stem(k, w)
plt.title('Сигнал $w[k]$')
plt.xlabel('$k$')
plt.ylabel('$w[k]$')
plt.tight_layout()
plt.grid()

plt.figure(figsize=[8, 4]);
nu, W = dtft(w, M=2048)
plt.plot(nu, abs(W));
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.grid()
```

```python
def deriv(x):
    return abs(1j/(2*np.pi)*(np.pi*(N*np.cos(np.pi*N*x)/np.sin(np.pi*x) - np.cos(np.pi*x)*np.sin(np.pi*N*x)/(np.sin(np.pi*x))**2)*np.exp(-1j*(N-1)*np.pi*x) + np.sin(np.pi*N*x)/np.sin(np.pi*x)*np.exp(-1j*(N-1)*np.pi*x) * (1-N)*1j*np.pi))

M=2048
nu = np.linspace(-1,1,M)
X = [deriv(nu1) for nu1 in nu]

plt.figure(figsize=[8, 4])
plt.plot(nu, X)
plt.xlim([-0.5, 0.5])
plt.ylim(bottom=0)
plt.title('ДВПФ w[k] (модуль), один период(Теоретический)')
plt.xlabel('$\\nu$')
plt.ylabel('$|W(\\nu)|$')
plt.grid()  
```

Теорема о дифференцировании работает.

## Task 2.3. Циклический сдвиг в ДПФ.

```python
y = np.ones(N)
for i in range(N):
    y[i] = x[(i - m) % N] 
    
plt.figure(figsize=[8, 3])
plt.stem(k, x)
plt.title('Сигнал $x[k]$')
plt.xlabel('$k$')
plt.ylabel('$x[k]$')
plt.tight_layout()
plt.grid()

plt.figure(figsize=[8, 3])
plt.stem(k, y)
plt.title('Сигнал $y[k]$')
plt.xlabel('$k$')
plt.ylabel('$y[k]$')
plt.tight_layout()
plt.grid()
```

```python
plt.figure(figsize=[8, 16])

n=np.arange(N)
Xn = scipy.linalg.dft(N) @ x
Yn = [Xn[i] * np.exp(-1j * 2 * np.pi * m * i / 8) for i in range(N)]
plt.subplot(4, 1, 1)
plt.stem(n, Xn.real)
plt.xticks(n)
plt.title('ДПФ сигнала $x[k]$')
plt.xlabel('$n$')
plt.ylabel('Re $X[n]$')
plt.grid()

plt.subplot(4, 1, 2)
plt.stem(n, Xn.imag)
plt.xticks(n)
plt.xlabel('$n$')
plt.ylabel('Im $X[n]$')
plt.grid()

Yn_real = [Yn[i].real for i in range(N)]
Yn_img = [Yn[i].imag for i in range(N)]

plt.subplot(4, 1, 3)
plt.stem(n, Yn_real)
plt.xticks(n)
plt.title('ДПФ сигнала $y[k]$')
plt.xlabel('$n$')
plt.ylabel('Re $Y[n]$')
plt.grid()

plt.subplot(4, 1, 4)
plt.stem(n, Yn_img)
plt.xticks(n)
plt.xlabel('$n$')
plt.ylabel('Im $Y[n]$')
plt.grid()

#plt.tight_layout;
```
